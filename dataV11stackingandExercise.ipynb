### horizontal and vertical stacking


```python
import numpy as np
```

```python
matrix1 = np.random.randint(1,20,9).reshape(3,3)
```

```python
matrix2 = np.random.randint(1,25,9).reshape(3,3)
```

```python
matrix1
```

```python
matrix2
```

```python
#horizonatal stacking
np.hstack((matrix1,matrix2))
```

```python
#vertical stacking
np.vstack((matrix1,matrix2))
```

```python
#note :- if we perform vertical stacking on 1D array then it becomes a 2D array
```
