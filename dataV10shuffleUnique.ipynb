### shuffle and unique 

```python
import numpy as np
```

```python
np.random.seed(100)
```

```python
arr = np.random.randint(1,15,15)
arr
```

```python
np.random.shuffle(arr)
arr
```

```python
np.unique(arr)
```

```python
np.unique(arr).size
```

```python
np.unique(arr, return_index=True,return_counts=True)
#returns index of unique numbers and count of times they occured
```
