#### data slicing 

```python
import numpy as np
```

```python
arr = np.array([4,54,65,38,34,85,72])
arr
```

```python
arr[::-1]
```

```python
arr[4:5]
```

```python
arr[-3:-1]
```

```python
arr[2:-1]
```

```python
arr[::3]
#entire array but every third element
```

```python
arr[-3:-5]
#wrong since we have to move left to right not vice versa
```

```python
arr[4:2]
#wrong
```

```python
arr[3:-5]
#wrong
```
