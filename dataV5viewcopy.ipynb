#### view vs copy


```python
import numpy as np
```

```python
np.random.seed(34)
arr = np.random.choice(np.arange(1,50), size=13, replace=False)
arr
```

```python
arr_sample = arr[4:9]
arr
```

```python
#changing the samole array to see the view
arr_sample[:] = 0
arr_sample
```

```python
#arr will also have zero at the place ofthe view
arr
```

```python
####################creating copy instead ####################
np.random.seed(69)
new_arr = np.random.choice(np.arange(40,60),size=9,replace=False)
new_arr
```

```python
new_arr_sample = new_arr[3:7].copy()
new_arr_sample
```

```python
new_arr_sample[:]=0
new_arr_sample
```
