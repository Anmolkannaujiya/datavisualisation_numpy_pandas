### series sorting and list array convertion

```python
import pandas as pd
```

```python
dic = {1:'one',2:'two',3:'three',4:'four'}
```

```python
series = pd.Series(dic)
```

```python
series
```

```python
series.sort_index()
```

```python
series = series.sort_index(ascending=False)
series
```

```python
series.sort_values()
```

```python
series.sort_values(ascending=False)
```

```python
series.tolist()
#python list
```

```python
series.values
#numpy array
```
