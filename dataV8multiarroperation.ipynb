### operation on multiple array 

```python
import numpy as np
```

```python
arr1 = np.array([2,5,15,18])
```

```python
arr2 = np.array([6,8,3,8,])
```

```python
arr1+arr2
#shape needs to be same
```

```python
arr1*arr2
```

```python
arr1.dot(arr2)
```
